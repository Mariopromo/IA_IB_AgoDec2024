```
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos de California
california = fetch_california_housing()
data_frame = pd.DataFrame(data=california.data, columns=california.feature_names)
data_frame['Valor'] = california.target

# Seleccionar la variable independiente y la dependiente
X_var = data_frame[['MedInc']]  # Ingresos medios en la zona
y_var = data_frame['Valor']

# Dividir los datos en conjuntos de entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train_set, X_test_set, y_train_set, y_test_set = train_test_split(X_var, y_var, test_size=0.2, random_state=42)

# Crear y entrenar el modelo de regresión lineal
regresion_modelo = LinearRegression()
regresion_modelo.fit(X_train_set, y_train_set)

# Realizar predicciones con el conjunto de prueba
y_pred_set = regresion_modelo.predict(X_test_set)

# Calcular el error cuadrático medio
error_cuadratico_medio = mean_squared_error(y_test_set, y_pred_set)
print(f"Error Cuadrático Medio: {error_cuadratico_medio}")

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(X_test_set, y_test_set, color='green', label='Datos originales')
plt.plot(X_test_set, y_pred_set, color='orange', linewidth=2, label='Predicciones del modelo')
plt.xlabel('Ingresos medios en la zona (MedInc)')
plt.ylabel('Valor de la casa')
plt.title('Regresión Lineal: Valor de la casa vs Ingresos medios en la zona')
plt.legend()
plt.show()
```
